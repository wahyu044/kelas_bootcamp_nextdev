```python
# dataframe
nilai_uts = {
    'nur' : 80,
    'wahyu' : 85,
    'melliano' : 90,
    'hariyanto' :85
}

nilai_uts = pd.Series(nilai_uts)
nilai_uts
```

```python
nilai_uts_df = pd.DataFrame(nilai_uts)
nilai_uts_df
```

```python
nilai_uts_df = pd.DataFrame(nilai_uts, columns=['nilai_uts'])
nilai_uts_df
```

```python
nilai_uas = {
    'nur' : 90,
    'wahyu' : 85,
    'melliano' : 80,
    'hariyanto' :85,
    'wayy' : 90
}

nilai_uas = pd.Series(nilai_uas)
nilai_uas
```

```python
data_nilai = pd.DataFrame({'UTS':nilai_uts,
                            'UAS':nilai_uas})
data_nilai
```

```python
nilai_tugas = {
    'nur' : 85,
    'wahyu' : 90,
    'melliano' : 85,
    'hariyanto' :80,
    'wayy' : 85
}

nilai_tugas = pd.Series(nilai_tugas)
nilai_tugas
```

```python
data_nilai['Tugas'] = nilai_tugas
data_nilai
```

```python
data_nilai['Praktikum'] = [90, 85, 80, 85, 90]
data_nilai
```

```python
# querying / masking
df_2014 = df[df['year_id'] > 2014]
df_2014['team_id'].value_counts()
```

```python
kondisi = (df['team_id'] == 'NYK') & (df['opp_pts'] > 100) & (df['game_result'] == 'W')
df[kondisi]
```

```python
kondisi = df['team_id'].str.contains('NY')
df[kondisi]['team_id'].value_counts()
```

```python
# grouping / aggregating
df.groupby('team_id')['pts'].sum()
```
