```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
statistik = pd.read_excel("COVID-19 Indonesia.xlsx", sheet_name='Statistik Harian')
statistik.head()
```

```python
statistik.describe()
```

```python
statistik.info()
```

```python
kolom = ['Date', 'Dosis pertama (harian)', 'Dosis kedua (harian)']
data = statistik[kolom]
data.head()
```

```python
data.tail()
```

```python
data.dropna(how='all', inplace=True)
data.tail()
```

```python
data.head()
```

```python
data.fillna(0, inplace=True)
data.head()
```

```python
data.info()
```

```python
data.set_index('Date', inplace=True)
data.head()
```
